```python
import pandas as pd
import matplotlib.pyplot as plt

# 그래프 설정
plt.rcParams['figure.figsize'] = (5, 3)        # (가로,세로) 인치 단위
plt.rcParams['axes.unicode_minus'] = False    # 그래프 눈금 값에서 (-)숫자표시
plt.rcParams['lines.linewidth'] = 2            # 선 굵기
plt.rcParams['font.family'] = 'Malgun Gothic'  # 한글 폰트 사용  (AppleGothic)

##사용할 데이터를 읽음
df = pd.read_csv('최저시급.csv', encoding='cp949')
df = df.set_index('년도')
df
```

```python
## 최저시급과 소비자총물가지수 비교

import pandas as pd
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()
line1 = ax1.plot(df.columns, df.values[0], 'o', linestyle='solid', color='red', label='최저시급')
ax1.set_ylim(0, 10000)
ax1.set_ylabel('최저시급', color = 'black', rotation = 90)
ax1.set_xlabel('연도')
ax1.tick_params(axis='x', labelrotation=90)

ax2 = ax1.twinx()
ax2.set_ylim(60, 110)
ax2.set_ylabel('소비자물가총지수\n(2020=100))', color = 'black', rotation = 90)
line2 = ax2.plot(df.columns, df.values[3], 'v', linestyle='solid', label='소비자물가총지수',)

lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels)
plt.title("연도별 최저시급과 소비자물가총지수")

plt.show()
```

```python
## 최저시급 미만율

import pandas as pd
import matplotlib.pyplot as plt

plt.plot(df.columns, df.values[2], 'o', linestyle='solid', label='최저시급 미만율', )

plt.title("최저시급 미만율")
plt.xlabel('연도')
plt.xticks(rotation = 90)
plt.ylabel('미만율')
plt.legend()

plt.show()
```
